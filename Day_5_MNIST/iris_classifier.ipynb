```python
import tensorflow as tf
import numpy as np

#create a linear model to find the difference between three different flowers

#iris setosa
#iris veriscolour
#iris virginica

from tensorflow.contrib.learn.python.learn.datasets import base

IRIS_TRAINING = "./iris_training.csv"
IRIS_TEST = "./iris_test.csv"
#training set is taking from the iris_training.csv file features are stored in a separate data set along with the target dtype
training_set = base.load_csv_with_header(filename=IRIS_TRAINING, features_dtype = np.float32, target_dtype=np.int)

test_set = base.load_csv_with_header(filename=IRIS_TEST, features_dtype = np.float32, target_dtype=np.int)

print(test_set.data)
print(test_set.target)
```

```python
feature_name = "flower_features"

feature_columns = [tf.feature_column.numeric_column(feature_name, shape=[4])]

classifier = tf.estimator.LinearClassifier(feature_columns = feature_columns, n_classes = 3, model_dir="./iris_model")

def input_fn(dataset):
    def _fn():
        features={feature_name: tf.constant(dataset.data)}
        label = tf.constant(dataset.target)
        return features, label
    return _fn

def eval_input_fn(features, batch_size):
    # Convert the inputs to a Dataset.
    dataset = tf.data.Dataset.from_tensor_slices(features)
    # Return the dataset.
    return dataset


#takes the data and sorts out features and labels into model
```

```python
classifier.train(input_fn = input_fn(test_set), steps = 1000)
```

```python
accuracy_score =  classifier.evaluate(input_fn(test_set), steps=100)["accuracy"]
predict = classifier.predict(input_fn(test_set))
expected = ['Setosa', 'Versicolor', 'Virginica']

print("\nAccuracy: {0:f}".format(accuracy_score))

for p in predict:
    class_id = p['class_ids'][0]
    probability = p['probabilities'][class_id]
    print('\nPrediction is "{}"'.format(p))
```
